```python
class User:
    def __init__(self):
        self.client_details_list = []
        self.loggedin = False
        self.cash = 1000
        self.TranferCash = False
        self.bank_name = None
        self.creation_date = datetime.datetime.now().strftime('%Y-%m-%d %H:%M:%S')
        self.last_used_date = None


# Registration
    def register(self, name, ph, pin, user_id, bank_name):
        cash = self.cash
        conditions = True
        if len(str(ph)) >= 20:
            print("Invalid Phone number!")
            conditions = False

        if len(pin) < 4 or len(pin) > 4:
            print("Enter 4-digit pin ")
            conditions = False

        if conditions == True:
            print("Account created successfully")
            self.client_details_list = [name, ph, pin, user_id, cash, bank_name]
            with open(f"{name}.csv", "w") as f:
                for details in self.client_details_list:
                    f.write(str(details) + "\n")

# Login
    def login(self, name, pin, user_id):
        with open(f"{name}.csv", "r") as f:
            details = f.read()
            self.client_details_list = details.split("\n")
            if len(self.client_details_list) > 2:
                self.loggedin = True
                self.last_used_date = datetime.datetime.now().strftime('%Y-%m-%d %H:%M:%S')
                print(f"{name} logged in")
                self.cash = int(self.client_details_list[4])
                self.name = name
                self.user_id = user_id
                self.pin = pin
            else:
                print("Wrong details")
                
# Deposit
    def add_cash(self, amount):
        if amount > 0:
            self.cash += amount
            with open(f"{name}.csv","r") as f:
                details = f.read()
                self.client_details_list = details.split("\n")
            
            with open(f"{name}.csv","w") as f:
                f.write(details.replace(str(self.client_details_list[4]),str(self.cash)))

            with open(f"{name}.csv", "a") as f:
                current_datetime = datetime.datetime.now()
                transaction_info = (f"\n{current_datetime.strftime('%Y-%m-%d %H:%M:%S')} - Deposited ${amount}")
                f.write(transaction_info + "\n")

            print("Amount added successfully")

        else:
            print("Enter correct value of amount")

# Withdraw
    def withdraw(self, amount):
        if amount < self.cash:
            self.cash -= amount
            with open(f"{name}.csv","r") as f:
                details = f.read()
                self.client_details_list = details.split("\n")
            
            with open(f"{name}.csv","w") as f:
                f.write(details.replace(str(self.client_details_list[4]),str(self.cash)))

            with open(f"{name}.csv", "a") as f:
                current_datetime = datetime.datetime.now()
                transaction_info = (f"\n{current_datetime.strftime('%Y-%m-%d %H:%M:%S')} - Withdrawn ${amount}")                  
                f.write(transaction_info + "\n")
      
            print("Amount withdraw successfully")
            print("Balacne left =", self.cash)
        
        else:
            print("Not enough balance")

        
# Transfer
    def Tranfer_cash(self, amount, name, ph):
        with open(f"{self.name}.csv", "r") as f:
            details = f.read()
            self.client_details_list = details.split("\n")
        if amount > self.cash:
            print("Not enough balance")
            return

        # Transfer the money
        with open(f"{name}.csv", "r") as f:
            details = f.read()
            self.client_details_list = details.split("\n")  
            total_cash = int(self.client_details_list[4]) + amount

        with open(f"{name}.csv", "w") as f:
            f.write(details.replace(str(self.client_details_list[4]), str(total_cash)))

        with open(f"{self.name}.csv", "r") as f:
            details = f.read()
            self.client_details_list = details.split("\n")
            left_cash = int(self.client_details_list[4]) - amount

        with open(f"{self.name}.csv", "w") as f:
            f.write(details.replace(str(self.client_details_list[4]), str(left_cash)))

        # Update the transaction history of the sender and receiver
        with open(f"{self.name}.csv", "a") as f:
            current_date = datetime.datetime.now()
            current_time = current_date.strftime("%H:%M:%S")
            current_date = current_date.strftime("%Y-%m-%d")
            f.write(f"\n{current_date} - {current_time} - Transferred ${amount} to {name} - {ph}")

        with open(f"{name}.csv", "a") as f:
            current_date = datetime.datetime.now()
            current_time = current_date.strftime("%H:%M:%S")
            current_date = current_date.strftime("%Y-%m-%d")
            f.write(f"\n{current_date} - {current_time} - Received ${amount} from {self.name}")
 
        print("Amount Transfered Successfully to", name)
        print("Balacne left =", left_cash)


# Transaction History
    def transaction_history(self):
        with open(f"{self.name}.csv", "r") as f:
            details = f.readlines()[6:][::-1]
            print("Transaction History: ")
            for detail in details:
                print(detail)

# Account Details
    def account_details(self):
        print("Account Details:")
        print(f"Name: {self.name}")
        print(f"PIN: {self.pin}")
        print(f"Balance: {self.cash}")
        print(f"Account Creation Date: {self.creation_date}")
        if self.last_used_date:
            print(f"Last Used Date: {self.last_used_date}")

        
# Piin Change
    def pin_change(self, pin):
        if len(pin) < 4 or len(pin) > 4:
            print("Enter 4-digit pin")
        else:
            with open(f"{self.name}.csv","r") as f:
                details = f.read()
                self.client_details_list = details.split("\n")

            with open(f"{self.name}.csv","w") as f:
                f.write(details.replace(str(self.client_details_list[2]),str(pin)))
            print("New Pin set up successfully")
        
# Phone Change
    def ph_change(self , ph):
        if len(str(ph)) > 10 or len(str(ph)) < 10:
            print("Invalid Phone number!")
        else:
            with open(f"{self.name}.csv","r") as f:
                details = f.read()
                self.client_details_list = details.split("\n")

            with open(f"{self.name}.csv","w") as f:
                f.write(details.replace(str(self.client_details_list[1]),str(ph)))
            print("New Phone number set up successfully")
            
# Banks
class MeezanBank(User):
    def __init__(self, name):
        super().__init__()  

class AlHabibBank(User):
    def __init__(self, name):
        super().__init__()  

# Check if the user entered a valid bank name
User_object = User()

# Check if the user entered a valid bank name
print("Welcome to the bank registration system!")
while True:
    bank_name = input("Select Bank: Meezan or Al-Habib: ")
    if bank_name in ["Meezan", "Al-Habib"]:
        break
    else:
        print("Invalid bank name. Please select either Meezan or Al-Habib.")


# Create a new bank object based on the user's selection
if bank_name == "Meezan":
    bank = MeezanBank(bank_name)
elif bank_name == "Al-Habib":
    bank = AlHabibBank(bank_name)

print("1.Login")
print("2.Create a new Account")
user = int(input("Make decision: "))

if user == 1:
    print("Logging in")
    name = input("Enter Name: ")
    user_id = input("Enter User-ID: ")
    pin = input("Enter Pin: ")
    User_object.login(name, user_id, pin)

    while True:
        if User_object.loggedin:
            print("1.Check Balcane")
            print("2.Deposit")
            print("3.Withdraw")
            print("4.Tranfer amount")
            print("5.Edit profile")
            print("6.Transaction History")
            print("7.Account Details")
            print("8.Logout")
            login_user = int(input())
            
            if login_user == 1:
                print("Balance =", User_object.cash)
                print("\n1.Return")
                print("2.Logout")
                choose = int(input())
                if choose == 1:
                    continue
                elif choose == 2:
                    break
                
            elif login_user == 2:
                print("Balance =", User_object.cash)
                amount = int(input("Enter amount: "))
                User_object.add_cash(amount)
                print("\n1.Return")
                print("2.Logout")
                choose = int(input())
                if choose == 1:
                    continue
                elif choose == 2:
                    break

            elif login_user == 3:
                print("Balance =", User_object.cash)
                amount = int(input("Enter amount: "))
                User_object.withdraw(amount)
                print("\n1.Return")
                print("2.Logout")
                choose = int(input())
                if choose == 1:
                    continue
                elif choose == 2:
                    break

            elif login_user == 4:
                print("Balance =", User_object.cash)
                amount = int(input("Enter amount: "))
                if amount < 0:
                    print("Enter a positive amount")
                    continue
                elif amount > User_object.cash:
                    print("Not enough balance")
                    continue
                name = input("Enter person name: ")
                ph = input("Enter person phone number: ")

                # Transfer the money
                User_object.Tranfer_cash(amount, name, ph)
                print("\n1. Return")
                print("2. Logout")
                choose = int(input())
                if choose == 1:
                    continue
                elif choose == 2:
                    break

            
            elif login_user == 5:
                print("1.Pin change")
                print("2.Phone Number change")
                edit_profile = int(input())
                if edit_profile == 1:
                    new_pin = input("Enter new Pin: ")
                    User_object.pin_change(new_pin)
                    print("\n1.Return")
                    print("2.Logout")
                    choose = int(input())
                    if choose == 1:
                        continue
                    elif choose == 2:
                        break
                    elif edit_profile == 2:
                        new_ph = int(input("Enter new Phone Number: "))
                        User_object.ph_change(new_ph)
                        print("\n1.Return")
                        print("2.Logout")
                        choose = int(input())
                        if choose == 1:
                            continue
                        elif choose == 2:
                            break

            elif login_user == 6:
                User_object.transaction_history()
                print("\n1.Return")
                print("2.Logout")
                choose = int(input())
                if choose == 1:
                    continue
                elif choose == 2:
                    break

            elif login_user == 7:
                User_object.account_details()
                print("\n1.Return")
                print("2.Logout")
                choose = int(input())
                if choose == 1:
                    continue
                elif choose == 2:
                    break
                    
            elif login_user == 8:
                break
                
                                
                
if user == 2:
    print("Creating a new Account")
    name = input("Enter Name: ")
    ph = int(input("Enter Phone Number: "))
    pin = input("Enter Pin: ")
    user_id = input("Enter User-ID: ")
    User_object.register(name, ph, pin, user_id, bank_name)

```

```python
class Bank(User):
    def __init__(self, bank_name, user_details):
        super().__init__()
        self.bank_name = bank_name
        self.client_details_list = user_details

    def get_bank_users_details(self):
        print(f"Users Details in {self.bank_name} Bank:")
        for user in self.client_details_list:
            print(f"Name: {user[0]} - Balance: {user[4]}")

    def total_bank_balance(self):
        total_balance = 0
        for user in self.client_details_list:
            total_balance += user[4]
        print(f"Total Money in {self.bank_name} Bank: {total_balance}")

user_details = [
    ["Mohsin Irfan", "3124192728", "2728", "22", 7000, "Meezan"],
    ["Harris Irfan", "3124192729", "2729", "23", 15000, "Al-Habib"],
    ["Mohsin Irfan", "3124192730", "2730", "24", 7000, "Meezan"],
    ["Aliza Shah", "3124192731", "2731", "25", 1000, "Meezan"],
]

print("Welcome to the bank registration system!")
bank_name = input("Select Bank: Meezan or Al-Habib : ")
if bank_name not in ["Meezan", "Al-Habib"]:
    print("Invalid bank name. Please try again.")
    exit()

if bank_name == "Meezan":
    bank = Bank(bank_name, [user for user in user_details if user[5] == "Meezan"])
elif bank_name == "Al-Habib":
    bank = Bank(bank_name, [user for user in user_details if user[5] == "Al-Habib"])

print(f"Welcome to {bank_name} Bank")

while True:
    print("1. Get Users Details")
    print("2. Total Bank Balance")
    print("3. Exit")
    choice = int(input("Enter your choice: "))

    if choice == 1:
        bank.get_bank_users_details()
        print("\n1.Return")
        print("2.Logout")
        choose = int(input())
        if choose == 1:
            continue
        elif choose == 2:
            break

    elif choice == 2:
        bank.total_bank_balance()
        print("\n1.Return")
        print("2.Logout")
        choose = int(input())
        if choose == 1:
            continue
        elif choose == 2:
            break
            
    elif choice == 3:
        break
    else:
        print("Invalid choice. Please try again.")
```
